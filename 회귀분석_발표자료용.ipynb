```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
# 합성 데이터 생성
np.random.seed(0)
X = np.linspace(0, 10, 100).reshape(-1, 1)  # 0부터 10까지의 100개 데이터 포인트
Y_linear = 2 * X.ravel() + 1  # 선형 구성 요소 (단순 선형 관계)
Y_nonlinear = Y_linear + 10 * np.sin(X.ravel()) + np.random.normal(0, 5, 100)  # 비선형 구성 요소

# 단순 선형 회귀 모델 적용
linear_model = LinearRegression()
linear_model.fit(X, Y_nonlinear)
Y_pred_linear = linear_model.predict(X)

# 복잡한 모델 적용 (다항 회귀)
poly_model = make_pipeline(PolynomialFeatures(degree=4), LinearRegression())
poly_model.fit(X, Y_nonlinear)
Y_pred_poly = poly_model.predict(X)

# 데이터 및 모델 시각화 - 선형 모델
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # 첫 번째 그래프 (선형 모델)
plt.scatter(X, Y_nonlinear, label='선형 및 비선형 구성 요소 데이터')
plt.plot(X, Y_pred_linear, color='red', label='선형 모델 적합')
plt.title("단순하지만 직관적인 모델")
plt.xlabel("X")
plt.ylabel("Y")
# plt.legend()
# plt.grid(True)

# 데이터 및 모델 시각화 - 다항 모델
plt.subplot(1, 2, 2)  # 두 번째 그래프 (다항 모델)
plt.scatter(X, Y_nonlinear, label='선형 및 비선형 구성 요소 데이터')
plt.plot(X, Y_pred_poly, color='green', label='다항 모델 적합')
plt.title("유연하지만 복잡한 모델")
plt.xlabel("X")
plt.ylabel("Y")
# plt.legend()
# plt.grid(True)

plt.show()
```
